# This Notebook is the main file that:
<ol>
  <li> Downloads the Figshare brain tumor dataset from `https://figshare.com/articles/dataset/brain_tumor_dataset/1512427` </li>
  <li> Unzips the downloaded data.</li>
  <li> Moves unzipped files to a specific directory called figshare_data.</li>
  <li> Converts .mat tumor files into numpy arrays</li>
  <li> Preprocesses tumor data (numpy arrays) by :</li>
  <ul>
      <li>Normalizing the tumor image pixels.</li>
      <li>Enhancing the tumor region using CLAHE algorithm.</li>
  </ul>
  <li> Partitions the data using Dirichlet distribution among <em>N</em> clients. </li>
  <li> Works with the following data to train the teacher models in different scenarios so that their ensemble knowledge is distilled into a simple student model: </li>
  <ul>
      <li>Brain Tumor Image</li>
      <li>Brain Tumor Labels</li>
      <li>Brain Tumor Region Masks</li>
  </ul>
</ol>

#### Public and test data

```python
public_data, public_labels = create_public_dataset(final_images, labels)
x_test, y_test = create_test_dataset(final_images, labels)
```

# Training and Evaluation

#### 2 Teachers (IID)

```python
num_clients = 2
num_rounds = 10

clients_data, clients_labels = partition_data(iid = True, num_clients = num_clients, tumor_images = final_images, tumor_labels = labels)

client_models_2 = [create_teacher_model() for _ in range(num_clients)]
student = create_student_model()

upload_cost_KD = 0
upload_cost_FL = 0
download_cost_FL = 0
download_cost_KD = 0

all_rounds_test_accuracy_2 = []
all_rounds_student_loss_2 = []

for round_num in range(num_rounds):
    print(f"Round {round_num + 1}/{num_rounds}")

    all_teacher_preds = []

    for client_idx in range(num_clients):
        print(f"Client {client_idx + 1}/{num_clients}")
        client_x, client_y = clients_data[client_idx], clients_labels[client_idx]
        model = client_models_2[client_idx]
        model.compile(optimizer=tf.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        model.fit(client_x, client_y, epochs=5, verbose=1)

        teacher_preds = model.predict(public_data)
        all_teacher_preds.append(teacher_preds)

        if round_num == 0: # upload cost for a teacher's soft labels as well as a teacher's parameters for one round only
          upload_cost_KD += teacher_preds.nbytes
          upload_cost_FL += sum([tf.size(var).numpy() * var.dtype.size for var in model.trainable_variables])

    aggregated_teacher_preds = np.mean(all_teacher_preds, axis=0)

    distiller = Distiller(student=student, teacher=model)
    distiller.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
        metrics=[tf.keras.metrics.SparseCategoricalAccuracy()],
        student_loss_fn=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
        distillation_loss_fn=tf.keras.losses.KLDivergence(),
        alpha=0.1,
        temperature=10,
    )

    history = distiller.fit(public_data, public_labels, epochs=5, verbose=1)

    training_accuracy = history.history['sparse_categorical_accuracy'][-1]

    results = distiller.evaluate(x_test, y_test, verbose=1)

    test_accuracy = results[distiller.metrics_names.index('sparse_categorical_accuracy')]
    all_rounds_test_accuracy_2.append(test_accuracy)
    all_rounds_student_loss_2.append(history.history['student_loss'][-1])
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Training Accuracy: {training_accuracy:.4f}")

# download cost for the student model (Knowledge Distillation)
download_cost_KD = np.sum([tf.size(var).numpy() * var.dtype.size for var in student.trainable_variables])

# download cost for averaged teachers parameters (Without Knowledge Distillation)
download_cost_FL = sum([tf.size(var).numpy() * var.dtype.size for var in client_models_2[0].trainable_variables])

communication_cost_2 = {
    'upload_cost_KD': upload_cost_KD,
    'download_cost_KD': download_cost_KD,
    'upload_cost_FL': upload_cost_FL,
    'download_cost_FL': download_cost_FL
    }


test_accuracy_df = pd.DataFrame(all_rounds_test_accuracy_2, columns=['Test Accuracy'])
test_accuracy_df.to_csv('all_rounds_test_accuracy_2.csv', index=False)

communication_cost_df = pd.DataFrame([communication_cost_2])
communication_cost_df.to_csv('communication_cost_2.csv', index=False)

student_loss_df = pd.DataFrame(all_rounds_student_loss_2, columns=['Student Loss'])
student_loss_df.to_csv('all_rounds_student_loss_2.csv', index=False)
```
