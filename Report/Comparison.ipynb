#### This Notebook plots the accuracy and the loss of the student model in different settings

#### Importing Dependencies

```python
from matplotlib import pyplot as plt
import pandas as pd
import os
import numpy as np
```

#### Constants

```python
num_rounds = 10
x_axis = [i for i in range(num_rounds)]
```

#### Non-IID setting (Student Accuracy)

```python
os.chdir("non-iid with 2 teachers/")
y_2teachers = pd.read_csv("all_rounds_test_accuracy_n2.csv")

os.chdir("../non-iid with 5 teachers/")
y_5teachers = pd.read_csv("all_rounds_test_accuracy_n5.csv")

os.chdir("..")


plt.figure(figsize=(15, 9))
plt.plot(x_axis, y_2teachers, marker = "o", ms = 7.5)
plt.plot(x_axis, y_5teachers, marker = "o", ms = 7.5)
plt.xticks(np.arange(len(y_2teachers)), np.arange(1, len(y_2teachers)+1))
plt.grid()
plt.legend(("2 Teachers", "5 Teachers"))
plt.xlabel("Communication Round")
plt.ylabel("Student Test Accuracy")
plt.show()
```

#### IID setting (Student Accuracy)

```python
os.chdir("iid with 2 teachers/")
y_2teachers = pd.read_csv("all_rounds_test_accuracy_2.csv")

os.chdir("../iid with 5 teachers/")
y_5teachers = pd.read_csv("all_rounds_test_accuracy_5.csv")

os.chdir("..")


plt.figure(figsize=(15, 9))
plt.plot(x_axis, y_2teachers, marker = "o", ms = 7.5)
plt.plot(x_axis, y_5teachers, marker = "o", ms = 7.5)
plt.xticks(np.arange(len(y_2teachers)), np.arange(1, len(y_2teachers)+1))
plt.grid()
plt.legend(("2 Teachers", "5 Teachers"))
plt.xlabel("Communication Round")
plt.ylabel("Student Test Accuracy")
plt.show()
```

#### Non-IID setting (Student Loss)

```python
os.chdir("non-iid with 2 teachers/")
y_2teachers = pd.read_csv("all_rounds_student_loss_n2.csv")

os.chdir("../non-iid with 5 teachers/")
y_5teachers = pd.read_csv("all_rounds_student_loss_n5.csv")

os.chdir("..")


plt.figure(figsize=(15, 9))
plt.plot(x_axis, y_2teachers, marker = "o", ms = 7.5)
plt.plot(x_axis, y_5teachers, marker = "o", ms = 7.5)
plt.xticks(np.arange(len(y_2teachers)), np.arange(1, len(y_2teachers)+1))
plt.grid()
plt.legend(("2 Teachers", "5 Teachers"))
plt.xlabel("Communication Round")
plt.ylabel("Student Loss")
plt.show()
```
